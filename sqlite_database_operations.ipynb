## Looking For Missingness

```python
df.columns
```

```python
import re
def cleaning_columns (df):
    def clean_name(name):
        cleaned_name = re.sub(r'[^a-zA-Z0-9]', '', name)
        return cleaned_name.lower()

    df.columns = [clean_name(col) for col in df.columns]
    return df

df = cleaning_columns(df)
df
```

```python
df.columns
```

### Dropping Missing Values

```python
df.dropna(subset=['cpthcpc','ndc','packagesize','packageunit','packagedescription','chargequantity'],inplace=True)
df
```
